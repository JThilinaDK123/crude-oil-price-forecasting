```python
def generate_calendar(year, drop_index=False):
    '''
    Simple function to generate a calendar containing
    US holidays, weekdays and  holiday weeks.
    '''
    from pandas.tseries.offsets import YearEnd
    from pandas.tseries.holiday import USFederalHolidayCalendar

    start_date = pd.to_datetime('1/1/'+str(year))
    end_date = start_date + YearEnd()
    DAT = pd.date_range(str(start_date), str(end_date), freq='D')
    MO = [d.strftime('%B') for d in DAT]
    holidays = USFederalHolidayCalendar().holidays(start=start_date, end=end_date)

    cal_df = pd.DataFrame({'date':DAT, 'month':MO})
    cal_df['year'] = [format(d, '%Y') for d in DAT]
    cal_df['weekday'] = [format(d, '%A') for d in DAT]
    cal_df['is_weekday'] = cal_df.weekday.isin(['Monday','Tuesday','Wednesday','Thursday','Friday'])
    cal_df['is_weekday'] = cal_df['is_weekday'].astype(int)
    cal_df['is_holiday'] = cal_df['date'].isin(holidays)
    cal_df['is_holiday'] = cal_df['is_holiday'].astype(int)
    cal_df['is_holiday_week'] = cal_df.is_holiday.rolling(window=7,center=True,min_periods=1).sum()
    cal_df['is_holiday_week'] = cal_df['is_holiday_week'].astype(int)

    if not drop_index: cal_df.set_index('date', inplace=True)

    return cal_df

def make_calendars(year_list, drop_index):
    cal_df = pd.DataFrame()
    for year in year_list:
        cal_df = pd.concat([cal_df, generate_calendar(year, drop_index=drop_index)])
#         cal_df = cal_df.append(generate_calendar(year, drop_index=drop_index))
    return cal_df

year_list = [str(int(i)) for i in np.arange(2001, 2024)]
cal_df = make_calendars(year_list, drop_index=False)
cal_df = cal_df.reset_index()
cal_df
```

### Mean Replacement

```python
def recursive_filter_prices_mean(data, threshold_multiplier=3, max_iterations=100):
    
    # Create a copy of the input data to avoid modifying the original
    filtered_data_mean = data.copy()

    for _ in range(max_iterations):
        # Calculate the sample mean and standard deviation
        mean_value = filtered_data_mean['Crude_oil_price'].mean()
        std_deviation = filtered_data_mean['Crude_oil_price'].std()

        # Identify spikes using the given threshold
        spikes = (filtered_data_mean['Crude_oil_price'] - mean_value).abs() >= threshold_multiplier * std_deviation
        if not spikes.any():
            # If no spikes are found, break out of the loop
            break

        # Replace outliers with respective monthly mean values
        filtered_data_mean['month_mean'] = filtered_data_mean.groupby(filtered_data_mean['date'].dt.to_period("M"))['Crude_oil_price'].transform('mean')
        filtered_data_mean.loc[spikes, 'Crude_oil_price'] = filtered_data_mean.loc[spikes, 'month_mean']
        filtered_data_mean = filtered_data_mean.drop(columns=['month_mean'])

    return filtered_data_mean
```

```python
df_crude = df_crude.sort_values('date')

# Apply the recursive filter on prices
df_MEAN = recursive_filter_prices_mean(df_crude)
df_MEAN
```

```python
# Calculate mean and standard deviation
mean_price = df_MEAN['Crude_oil_price'].mean()
std_dev = df_MEAN['Crude_oil_price'].std()
print('New Mean Price after Mean Replacement: ', mean_price)
print('New Standard Deviation after Mean Replacement: ', std_dev)
```

```python
plot_df(df_MEAN, x=df_MEAN['date'], y=df_MEAN['Crude_oil_price'], 
        title='WTI Crude Oil Prices from 2000-01 to 2023-10 (RFP - Mean Value Replacements)')
```

### Median Replacement

```python
def recursive_filter_prices_median(data, threshold_multiplier=3, max_iterations=100):
    # Create a copy of the input data to avoid modifying the original
    filtered_data_median = data.copy()

    for _ in range(max_iterations):
        # Calculate the sample mean and standard deviation
        mean_value = filtered_data_median['Crude_oil_price'].mean()
        std_deviation = filtered_data_median['Crude_oil_price'].std()

        # Identify spikes using the given threshold
        spikes = (filtered_data_median['Crude_oil_price'] - mean_value).abs() >= threshold_multiplier * std_deviation
        if spikes.empty:
            # If no spikes are found, break out of the loop
            break

        # Replace Outliers
        filtered_data_median['month_mean'] = filtered_data_median.groupby(filtered_data_median['date'].dt.to_period("M"))['Crude_oil_price'].transform('median')
        filtered_data_median.loc[spikes, 'Crude_oil_price'] = filtered_data_median.loc[spikes, 'month_mean']
        filtered_data_median = filtered_data_median.drop(columns=['month_mean'])

    return filtered_data_median
```

```python
# Calculate mean and standard deviation
mean_price = df_MEDIAN['Crude_oil_price'].mean()
std_dev = df_MEDIAN['Crude_oil_price'].std()
print('New Mean Price after Median Replacement: ', mean_price)
print('New Standard Deviation after Median Replacement: ', std_dev)
```

```python
plot_df(df_MEDIAN, x=df_MEDIAN['date'], y=df_MEDIAN['Crude_oil_price'], 
        title='WTI Crude Oil Prices from 2000-01 to 2023-10 (RFP - Median Value Replacements)')
```

### Damping Replacement 

```python
def recursive_filter_prices_damping(data, threshold_multiplier=3,max_iterations=100):
    
    # Create a copy of the input data to avoid modifying the original
    filtered_data_damp = data.copy()

    for _ in range(max_iterations):
        # Calculate the sample mean and standard deviation
        mean_value = filtered_data_damp['Crude_oil_price'].mean()
        std_deviation = filtered_data_damp['Crude_oil_price'].std()

        # Identify spikes using the given threshold
        spikes = (filtered_data_damp['Crude_oil_price'] - mean_value).abs() >= threshold_multiplier * std_deviation
        if spikes.empty:
            # If no spikes are found, break out of the loop
            break

        # Replace Outliers
        Y1 = mean_value
        filtered_data_damp.loc[spikes, 'Crude_oil_price'] = Y1 + Y1 * np.log10(filtered_data_damp.loc[spikes, 'Crude_oil_price'] / Y1)

    return filtered_data_damp
```

```python
# Apply the recursive filter on prices
df_DAMPING = recursive_filter_prices_damping(df_crude)
df_DAMPING
```

```python
# Calculate mean and standard deviation
mean_price = df_DAMPING['Crude_oil_price'].mean()
std_dev = df_DAMPING['Crude_oil_price'].std()
print('New Mean Price after Damping Replacement: ', mean_price)
print('New Standard Deviation after Damping Replacement: ', std_dev)
```

```python
plot_df(df_DAMPING, x=df_DAMPING['date'], y=df_DAMPING['Crude_oil_price'], 
        title='WTI Crude Oil Prices from 2000-01 to 2023-10 (RFP - After Damping Scheme)')
```

### Threshold Replacement 

```python
def recursive_filter_prices_threshold(data, threshold_multiplier=3, max_iterations=100):
    # Create a copy of the input data to avoid modifying the original
    filtered_data_median = data.copy()

    for _ in range(max_iterations):
        # Calculate the sample mean and standard deviation
        mean_value = filtered_data_median['Crude_oil_price'].mean()
        std_deviation = filtered_data_median['Crude_oil_price'].std()

        # Identify spikes using the given threshold
        spikes = (filtered_data_median['Crude_oil_price'] - mean_value).abs() >= threshold_multiplier * std_deviation
        if not spikes.any():
            # If no spikes are found, break out of the loop
            break

        # Calculate the mean and median for each month
        monthly_stats = filtered_data_median.groupby(filtered_data_median['date'].dt.to_period("M"))['Crude_oil_price'].agg(['mean', 'median'])
        monthly_stats['mean_median'] = (monthly_stats['mean'] + monthly_stats['median']) / 2

        # Merge the stats back to the original data
        filtered_data_median = filtered_data_median.merge(monthly_stats[['mean_median']], left_on=filtered_data_median['date'].dt.to_period("M"), right_index=True, how='left')

        # Replace outliers
        filtered_data_median.loc[spikes, 'Crude_oil_price'] = filtered_data_median.loc[spikes, 'mean_median']

        # Drop the temporary columns
        filtered_data_median = filtered_data_median.drop(columns=['mean_median'])

    return filtered_data_median
```

```python
# Apply the recursive filter on prices
df_THRESHOLD = recursive_filter_prices_threshold(df_crude)
df_THRESHOLD = df_THRESHOLD.drop(columns = ['key_0'])
df_THRESHOLD
```

```python
# Calculate mean and standard deviation
mean_price = df_THRESHOLD['Crude_oil_price'].mean()
std_dev = df_THRESHOLD['Crude_oil_price'].std()
print('New Mean Price after threshold Replacement: ', mean_price)
print('New Standard Deviation after threshold Replacement: ', std_dev)
```

```python
plot_df(df_THRESHOLD, x=df_THRESHOLD['date'], y=df_THRESHOLD['Crude_oil_price'], 
        title='WTI Crude Oil Prices from 2000-01 to 2023-10 (RFP - After threshold Scheme)')
```

```python
df_THRESHOLD.to_csv('data/RFP_threshold_Replacements.csv' , index = False)
```
